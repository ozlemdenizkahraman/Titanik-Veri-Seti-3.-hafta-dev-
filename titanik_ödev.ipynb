```python
import pandas as pd

file_path = '/content/drive/MyDrive/Colab Notebooks/Titanic_dataset.csv'  # Adjust if your file name is different
df_titanic = pd.read_csv(file_path)

df_titanic.head()
```

```python
#kazada ölenlerin yaş ortalaması
olen_yas = df_titanic[df_titanic['Survived'] == 0]['Age'].mean()
print("Ölen İnsan Yaş Ortalaması: ", olen_yas)
```

```python
# Kazada ölenlerin bilet fiyatlarının ortalamasını ve medyanı
kaza_olenler = df_titanic[df_titanic['Survived'] == 0]
olen_fiyat_ortalama = kaza_olenler['Fare'].mean()
olen_fiyat_medyan = kaza_olenler['Fare'].median()
print("Ortalama:", olen_fiyat_ortalama, " Medyan: ", olen_fiyat_medyan)
```

```python
#Ölen erkek yaş ortalama
olen_erkek = df_titanic[(df_titanic['Survived'] == 0) & (df_titanic['Sex'] == 'male')]['Age'].mean()
print("Ölen Erkek Yaş Ortalaması: ", olen_erkek)
```

```python
#Ölen kadın yaş ortalama
olen_kadin = df_titanic[(df_titanic['Survived'] == 0) & (df_titanic['Sex'] == 'female')]['Age'].mean()
print("Ölen Kadın Yaş Ortalaması: ", olen_kadin)
```

```python
#Kazadan kurtulanların yaş ortalaması
kurtulan_yas = df_titanic[df_titanic['Survived'] == 1]['Age'].mean()
print("Kurtulan İnsan Yaş Ortalaması: ", kurtulan_yas)
```

```python
#Kazadan kurtulanların bilet fiyatlarının ortalaması
kurtulanlar_bilet_ortalama = df_titanic[df_titanic['Survived'] == 1]['Fare'].mean()
print("Kurtulanların bilet fiyat Ortalaması: ", kurtulanlar_bilet_ortalama)
```

```python
#Kazadan kurtulan toplam kişi sayısı
kurtulanlar_toplam_kişi = len(df_titanic[df_titanic['Survived'] == 1])
print("Kazadan kurtulan toplam kişi sayısı: ", kurtulanlar_toplam_kişi)
```

```python
#10 yaşından küçüklerin bilet fiyatlarının medyan değeri
on_yas_bilet_fiyat_ort = df_titanic[df_titanic['Age'] < 10]['Fare'].median()
print("10 yaş bilet fiyat ortalaması: ", on_yas_bilet_fiyat_ort)
```

```python
#1., 2. ve 3. sınıf bilet fiyatlarının ortalama ve medyanlarını karşılaştırınız
df_titanic.groupby('Pclass')['Fare'].agg(['mean', 'median'])
```

```python
#Kazada ölen kadınların oranı ile erkeklerin oranını karşılaştırınız
# Ölen kadın oranı
olen_kadin_orani = len(df_titanic[(df_titanic['Sex'] == 'female') & (df_titanic['Survived'] == 0)]) / len(df_titanic[df_titanic['Sex'] == 'female'])

# Ölen erkek oranı
olen_erkek_orani = len(df_titanic[(df_titanic['Sex'] == 'male') & (df_titanic['Survived'] == 0)]) / len(df_titanic[df_titanic['Sex'] == 'male'])

olen_kadin_orani, olen_erkek_orani
```
